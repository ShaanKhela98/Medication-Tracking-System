You are a health care team member working in a small clinic that manages a significant number of patients with chronic conditions, such as diabetes and hypertension. The clinic administrator has come to your team with an alarming issue. A significant number of patients are noncompliant in their medication schedules. Noncompliance can lead to worsening conditions, increased hospital visits, and overall poorer patient outcomes. To help manage this, your administrator wants to develop a simple Python-based system to track and remind patients about their medication schedules.

The purpose of this assignment is to develop a simple Python-based system that tracks medication schedules and provides reminders to patients.

Using Python script, perform the following tasks for seven patients with at least three medications:

- Store patient information (name, age, condition, medication list) in compliance with HIPAA regulations.
- Track medication schedules (name of medication, dosage, frequency, and time of day).
- Create a function that checks the current time against the medication schedule.
- Send reminders (simulated by print statements) to the patient when it is time to take their medication.
- Use loops to simulate time passing and trigger reminders.
- Generate reports on missed medications to help the care team follow up with patients.
- Track missed reminders and generate a report that lists all missed doses.

```python
#Store patient information (name, age, condition, medication list) in compliance with HIPAA regulations.

import pandas as pd

df = pd.read_csv('hipaa_compliant_patient_data.csv')
df
```

```python
#Track medication schedules (name of medication, dosage, frequency, and time of day).

df2 = pd.read_csv('medication_schedule.csv')
df2
```

```python
#Create a function that checks the current time against the medication schedule.

def check_medication_schedule(current_time, medication_schedule):
    reminders = []
    for index, row in medication_schedule.iterrows():
        if row['Time of Day'] == current_time:
            reminders.append(
                f"Patient {row['Patient ID']}: Time to take {row['Name of Medication']} at {row['Time of Day']} with dosage {row['Dosage']}."
            )
    return reminders
```

```python
#Send Reminders (simulated by print statements) to the patient when it is time to take their medication.
def send_reminders(reminders):
    for reminder in reminders:
        print(reminder)
```

```python
# Use loops to simulate time passing and trigger reminders.
import time
from datetime import datetime

# Get current system time in HH:MM format
system_time = datetime.now().strftime("%H:%M")
print(f"Current System Time: {system_time}")

unique_times = sorted(df2['Time of Day'].unique())

# Create checked_times: all times up to and including the current system time
checked_times = [t for t in unique_times if t <= system_time]

print(f"Checked Times (up to system time): {checked_times}")

# Simulate time passing and trigger reminders for each checked time
for current_time in checked_times:
    print(f"\n--- Current Time: {current_time} ---")
    reminders = check_medication_schedule(current_time, df2)
    send_reminders(reminders)
    time.sleep(1)  # Simulate time passing 

# Generate reports on missed medications (not yet reminded)
missed_medications = df2[~df2['Time of Day'].isin(checked_times)]
missed_report = missed_medications[['Patient ID', 'Name of Medication', 'Dosage', 'Time of Day']]
print("\nMissed Medications Report:")
print(missed_report)

#Track missed reminders and generate a report that lists all missed doses.
missed_report.to_csv('missed_reminders_report.csv', index=False)
print("Missed reminders report saved to 'missed_reminders_report.csv'.")


```
